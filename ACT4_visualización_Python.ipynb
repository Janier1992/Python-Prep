```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/global_water.csv', encoding='latin-1', delimiter=';')
df.head()

df.head(100)
```

# Analisis Exploratorio de los datos

```python
print("\n Valores nulos por columna:") # validar valores nulos
null_counts = df.isnull().sum()
display(null_counts[null_counts > 0])
```

```python
print("\n Estadísticas descriptivas:") ## Medidas de tendencia central
display(df.describe(include='all'))
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns ## Medidas de dispersión, rango y distribuciones para columnas numericas
print("\n Rango de variables numéricas:")
for col in numeric_cols:
    print(f"\n{col}:")
    print(f"  Mínimo: {df[col].min():.2f}")
    print(f"  Máximo: {df[col].max():.2f}")
    print(f"  Media: {df[col].mean():.2f}")
    print(f"  Mediana: {df[col].median():.2f}")
    print(f"  Moda: {df[col].mode().values[0]:.2f}")
```

```python
plt.figure(figsize=(10, 4)) ## Grafico de valores nulos
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Mapa Valores Nulos")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df[numeric_cols], orient='v')
plt.title("Boxplots de Variables Numéricas")
plt.show()
```

```python
## Luego de hacer un analisis exploratorio de datased, se procederá a realizar algunos gráficos

print(df.columns.tolist())  # Mostrar todos los nombres de columnas
```

```python
# 1. Grafico de linea, Convertir la columna 'año' a tipo numérico:
df['año'] = pd.to_numeric(df['año'], errors='coerce')  # 'coerce' convierte los valores no numéricos a NaN

# 2. Convertir la columna 'consumo_total_de_agua_bmc' a tipo numérico:
df['consumo_total_de_agua_bmc'] = pd.to_numeric(df['consumo_total_de_agua_bmc'], errors='coerce')

# 3. Eliminar filas con valores faltantes (NaN) en las columnas relevantes:
df = df.dropna(subset=['año', 'consumo_total_de_agua_bmc'])

# 4. Graficar los datos:
df.groupby('año')['consumo_total_de_agua_bmc'].sum().plot(kind='line', color='blue', marker='o')
plt.title('Consumo Total de Agua por Año')
plt.xlabel('Año')
plt.ylabel('m³')
plt.grid()
plt.show()
```

```python
# 2. Uso por sector agrícola, industrial y doméstico

# 1. Datos de ejemplo (si tu DataFrame falla)
data = {
    'año': [2020, 2021, 2022],
    'uso_de_agua_agrícola': [40, 30, 50],
    'uso_de_agua_industrial': [30, 40, 20],
    'uso_de_agua_doméstico': [30, 30, 30]
}
df = pd.DataFrame(data)

# 2. Gráfico de barras apiladas (asegura que los datos son visibles)
df.set_index('año')[['uso_de_agua_agrícola', 'uso_de_agua_industrial', 'uso_de_agua_doméstico']]\
  .plot(kind='bar', stacked=True, figsize=(10, 6), color=['#2ecc71', '#3498db', '#f39c12'])

plt.title('Distribución del Uso de Agua por Año', pad=20)
plt.ylabel('Porcentaje (%)')
plt.xlabel('Año')
plt.legend(title='Sector', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('/content/global_water.csv', encoding='latin-1', delimiter=';')
df.head()

# Mostrar las primeras filas para entender la estructura
print(df.head())
print("\nInformación del dataset:")
print(df.info())
```

```python
# 3. Gráfico de barras del consumo medio de agua por país
plt.figure(figsize=(12, 6))
sns.barplot(x='Pais', y='consumo_total_de_agua_bmc', data=df, errorbar=None, palette='magma', hue='Pais', legend=False) # Changed df_viz to df
plt.title('Consumo Medio de Agua por País')
plt.xlabel('País')
plt.ylabel('Consumo Medio de Agua (bmc)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
# Comentario: Compara el consumo medio de agua entre países.
```

```python
plt.figure(figsize=(12, 6))
# Convert 'consumo_total_de_agua_bmc' to numeric, handling commas and errors
df['consumo_total_de_agua_bmc'] = pd.to_numeric(df['consumo_total_de_agua_bmc'].str.replace(',', '.'), errors='coerce')
df.groupby('Pais')['consumo_total_de_agua_bmc'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 países con mayor consumo promedio de agua (bmc)', fontsize=14)
plt.xlabel('País', fontsize=12)
plt.ylabel('Consumo promedio de agua (bmc)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2. Diagrama de dispersión del consumo total de agua vs. uso de agua por capital
plt.figure(figsize=(10, 6))
sns.scatterplot(x='consumo_total_de_agua_bmc', y='uso_de_agua_por_capital', data=df, hue='nivel_de_escasez_de_agua', palette='viridis') # Changed df_viz to df
plt.title('Consumo Total de Agua vs. Uso de Agua por Capital')
plt.xlabel('Consumo Total de Agua (bmc)')
plt.ylabel('Uso de Agua por Capital')
plt.show()
# Comentario: Explora la relación entre el consumo total y el uso por capital, coloreado por nivel de escasez.
```

```python
##  Evolución del consumo de agua a lo largo del tiempo-----------------
# Filtrar los datos para los países seleccionados
countries = ['Brazil', 'China', 'USA', 'India', 'Russia']
filtered_df = df[df['Pais'].isin(countries)]

# Crear el gráfico de líneas con Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_df, x='año', y='consumo_total_de_agua_bmc', hue='Pais', marker='o', palette='viridis')

# Personalizar el gráfico
plt.title('Evolución del consumo total de agua por país', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Consumo total de agua (bmc)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='best', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
## Distribución del nivel de escasez de agua (bajo-medio-alto/Low-Medium-High)

plt.figure(figsize=(8, 6))
df['nivel_de_escasez_de_agua'].value_counts().plot(kind='pie', autopct='%1.1f%%',
                                                  colors=['lightgreen', 'gold', 'lightcoral'],
                                                  explode=(0.05, 0.05, 0.05))
plt.title('Distribución de niveles de escasez de agua', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
## Relación entre precipitación anual y consumo doméstico

plt.figure(figsize=(10, 6))
plt.scatter(df['impacto_de_la_lluvia_precipitacion_anual_mm'],
            df['uso_de_agua_doméstico'],
            alpha=0.5, color='teal')
plt.title('Relación entre precipitación anual y uso doméstico de agua', fontsize=14)
plt.xlabel('Precipitación anual (mm)', fontsize=12)
plt.ylabel('Uso doméstico de agua (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
## Comparación de usos del agua (agrícola, industrial, doméstico)
plt.figure(figsize=(10, 6))
uses = ['uso_de_agua_agrícola', 'uso_de_agua_industrial', 'uso_de_agua_doméstico']

# Convert columns to numeric before calculating the mean
for col in uses:
    df[col] = pd.to_numeric(df[col].str.replace(',', '.'), errors='coerce')

means = df[uses].mean()
plt.bar(uses, means, color=['green', 'blue', 'purple'])
plt.title('Distribución promedio de usos del agua', fontsize=14)
plt.xlabel('Tipo de uso', fontsize=12)
plt.ylabel('Porcentaje promedio', fontsize=8)
plt.ylim(0, 50)  # Set y-axis limit to 50
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 4. Diagrama de caja del consumo de agua según el nivel de escasez
plt.figure(figsize=(10, 6))
sns.boxplot(x='nivel_de_escasez_de_agua', y='consumo_total_de_agua_bmc', data=df, palette='rocket') # Changed df_viz to df
plt.title('Consumo de Agua según el Nivel de Escasez')
plt.xlabel('Nivel de Escasez de Agua')
plt.ylabel('Consumo Total de Agua (bmc)')
plt.show()
# Comentario: Muestra la distribución del consumo de agua para cada nivel de escasez.
```

```python
## Correlación entre variables numéricas
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de correlación entre variables numéricas', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
## Relación entre uso agrícola e industrial por país
plt.figure(figsize=(12, 8))
top_countries = df['Pais'].value_counts().index[:10]  # Tomamos los 10 países con más datos
filtered_df = df[df['Pais'].isin(top_countries)]
sns.scatterplot(x='uso_de_agua_agrícola', y='uso_de_agua_industrial',
                hue='Pais', data=filtered_df, palette='tab10', s=100)
plt.title('Relación entre uso agrícola e industrial del agua por país', fontsize=14)
plt.xlabel('Uso agrícola (%)', fontsize=12)
plt.ylabel('Uso industrial (%)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
##  Distribución del uso de agua per cápita
plt.figure(figsize=(10, 6))
sns.histplot(df['uso_de_agua_por_capital'], bins=30, kde=True, color='royalblue')
plt.title('Distribución del uso de agua per cápita', fontsize=14)
plt.xlabel('Uso de agua per cápita', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
## Tendencias anuales del agotamiento de aguas subterráneas
plt.figure(figsize=(12, 6))
sns.lineplot(x='año', y='porcentaje_tasa_de_agotamiento_de_las_aguas_subterráneas',
             data=df, ci=None, color='darkorange', marker='o') # Changed 'ańo' to 'año'
plt.title('Tendencia anual de la tasa de agotamiento de aguas subterráneas', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Tasa de agotamiento (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
