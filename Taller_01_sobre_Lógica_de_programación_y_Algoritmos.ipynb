***Suma de n´umeros pares e impares:*** Programa que solicita al usuario
que ingrese 5 n´umeros y muestra la suma de los n´umeros pares y la suma
de los n´umeros impares que se han ingresado.

```python
import pandas as pd

# Solicitar al usuario que ingrese 5 números
numeros = []
for i in range(5):
    numero = int(input("Ingrese un número: "))
    numeros.append(numero)

# Inicializar las variables para almacenar la suma de números pares e impares
suma_pares = 0
suma_impares = 0

# Calcular la suma de números pares e impares
for numero in numeros:
    if numero % 2 == 0:
        suma_pares += numero
    else:
        suma_impares += numero

# Mostrar la suma de números pares e impares
print(f"La suma de los números pares es: {suma_pares}")
print(f"La suma de los números impares es: {suma_impares}")
```

***Tabla de multiplicar***:  Pide al usuario un n´umero de 1 a 9 y muestra la
tabla de multiplicar de ese n´umero del 1 al 10.

```python
# Solicitar al usuario que ingrese un número del 1 al 9
numero = int(input("Ingrese un número del 1 al 9: "))

# Validar que el número esté en el rango correcto
if numero < 1 or numero > 9:
    print("Número fuera de rango. Por favor, ingrese un número del 1 al 9.")
else:
    # Calcular y mostrar la tabla de multiplicar
    print(f"Tabla de multiplicar del {numero}:")
    for i in range(1, 11):
        resultado = numero * i
        print(f"{numero} x {i} = {resultado}")
```

***Factorial de un n´umero***: Solicita al usuario un n´umero entre 15 y 35.
Debe mostrar el factorial del numero ingresado sin usar funciones directas
de calculo.

```python
# Solicitar al usuario que ingrese un número entre 15 y 35
numero = int(input("Ingrese un número entre 15 y 35: "))

# Validar que el número esté en el rango correcto
if numero < 15 or numero > 35:
    print("Número fuera de rango. Por favor, ingrese un número entre 15 y 35.")
else:
    # Calcular el factorial del número
    factorial = 1
    for i in range(1, numero + 1):
        factorial *= i

    # Mostrar el factorial
    print(f"El factorial de {numero} es: {factorial}")
```

***Promedio de n´umeros***: Pide al usuario una serie de n´umeros separados
por comas y muestra su promedio.

```python
# Solicitar al usuario que ingrese una serie de números separados por comas
numeros = input("Ingrese una serie de números separados por comas: ")

# Separar los números ingresados
numeros_lista = numeros.split(',')

# Convertir los números a tipo float
numeros_float = [float(num) for num in numeros_lista]

# Calcular el promedio de los números
promedio = sum(numeros_float) / len(numeros_float)

# Mostrar el promedio
print(f"El promedio de los números ingresados es: {promedio}")
```

***Contador de palabras:*** Solicita al usuario una frase y cuenta cu´antas
palabras tiene repetidas.

```python
# Solicitar al usuario que ingrese una frase
frase = input("Ingrese una frase: ")

# Separar las palabras de la frase
palabras = frase.split()

# Crear un diccionario para contar las palabras
contador_palabras = {}
for palabra in palabras:
    # Convertir la palabra a minúsculas para evitar contar palabras iguales con diferente capitalización
    palabra = palabra.lower()
    if palabra in contador_palabras:
        contador_palabras[palabra] += 1
    else:
        contador_palabras[palabra] = 1

# Mostrar el resultado
print("Palabras repetidas en la frase:")
for palabra, cantidad in contador_palabras.items():
    if cantidad > 1:
        print(f"{palabra}: {cantidad} veces")
```

***Fibonacci hasta cierto l´ımite***: Pide al usuario un l´ımite menor a 100
y muestra la secuencia de Fibonacci hasta ese l´ımite.

```python
# Solicitar al usuario que ingrese un límite menor a 100
limite = int(input("Ingrese un límite menor a 100: "))

# Validar que el límite sea menor a 100
if limite >= 100:
    print("El límite debe ser menor a 100.")
else:
    # Inicializar la secuencia de Fibonacci
    fibonacci = [0, 1]

    # Calcular la secuencia de Fibonacci hasta el límite
    while fibonacci[-1] + fibonacci[-2] <= limite:
        fibonacci.append(fibonacci[-1] + fibonacci[-2])

    # Mostrar la secuencia de Fibonacci
    print("Secuencia de Fibonacci:")
    print(fibonacci)
```

***Verificador de palíndromos***: Pide al usuario una palabra y verifica si
es un pal´ındromo o no.

```python
# Solicitar al usuario que ingrese una palabra
palabra = input("Ingrese una palabra: ")

# Verificar si la palabra es un palíndromo
es_palindromo = palabra.lower() == palabra.lower()[::-1]

# Mostrar el resultado
if es_palindromo:
    print(f"{palabra} es un palíndromo.")
else:
    print(f"{palabra} no es un palíndromo.")
```

***Ordenamiento de lista***: Pide al usuario una lista de n´umeros separados
por comas y ord´enalos de menor a mayor.

```python
# Solicitar al usuario que ingrese una lista de números separados por comas
numeros = input("Ingrese una lista de números separados por comas: ")

# Separar los números de la lista
numeros_lista = numeros.split(',')

# Convertir los números a tipo float
numeros_float = [float(num) for num in numeros_lista]

# Ordenar los números de menor a mayor
numeros_ordenados = sorted(numeros_float)

# Mostrar la lista ordenada
print("Lista de números ordenada de menor a mayor:")
print(numeros_ordenados)
```
